```python
import pandas as pd
import numpy as np

# ---- CONFIGURATION ----
INITIAL_RATING = 890
BASE_K = 32
WEIGHTS = {
    'sig_reg_percent': 0.95,
    'strike_percent': 0.95,
    'sig_strike_attempts': 0.90,
    'strike_attempts': 0.90,
    'sig_strike_landed': 0.85,
    'strike_landed': 0.85
}

def calculate_comp_score(fight_stats):
    return sum(fight_stats.get(metric, 0) * weight for metric, weight in WEIGHTS.items())

def expected_score(rating_a, rating_b):
    return 1 / (1 + 10**((rating_b - rating_a) / 400))

def update_elo(rating_a, rating_b, comp_score_a, comp_score_b, outcome_a):
    k_factor_a = BASE_K * (0.5 + comp_score_a / 100)
    k_factor_b = BASE_K * (0.5 + comp_score_b / 100)
    e_a = expected_score(rating_a, rating_b)
    new_rating_a = rating_a + k_factor_a * (outcome_a - e_a)
    new_rating_b = rating_b + k_factor_b * ((1 - outcome_a) - (1 - e_a))
    return new_rating_a, new_rating_b

# ---- DATA PREPARATION ----
# Ensure required columns exist and fill NAs for robust processing
df['fighter'] = df['fighter'].fillna('UnknownFighter')
df['res'] = df['res'].fillna('N')  # Treat missing result as No Contest
df['fight_pk'] = df['fight_pk'].fillna(-1)  # Just in case

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.sort_values(by=['date', 'fight_pk']).reset_index(drop=True)
df['elo_after_fight'] = np.nan  # Initialize column

# ---- INITIALIZE RATINGS ----
fighter_ratings = {}
for fighter in df['fighter'].unique():
    fighter_ratings[fighter] = INITIAL_RATING

# ---- PROCESS FIGHTS ----
for fight_pk, fight_group in df.groupby('fight_pk'):
    indices = fight_group.index.tolist()
    fighters = fight_group['fighter'].tolist()
    pre_fight_ratings = {fighter: fighter_ratings.get(fighter, INITIAL_RATING) for fighter in fighters}

    # If not exactly 2 fighters, just assign pre-fight rating
    if len(fight_group) != 2:
        for idx in indices:
            fighter = df.at[idx, 'fighter']
            df.at[idx, 'elo_after_fight'] = pre_fight_ratings[fighter]
        continue

    row1, row2 = fight_group.iloc[0], fight_group.iloc[1]
    fighter1, fighter2 = row1['fighter'], row2['fighter']
    res1, res2 = row1['res'], row2['res']

    # Handle valid outcomes
    if (res1, res2) == ('W', 'L'):
        outcome = 1
    elif (res1, res2) == ('L', 'W'):
        outcome = 0
    elif (res1, res2) == ('D', 'D'):
        outcome = 0.5
    else:
        # No contest or unhandled: assign pre-fight rating
        for idx in indices:
            fighter = df.at[idx, 'fighter']
            df.at[idx, 'elo_after_fight'] = pre_fight_ratings[fighter]
        continue

    # Get ratings and stats
    r1 = fighter_ratings.get(fighter1, INITIAL_RATING)
    r2 = fighter_ratings.get(fighter2, INITIAL_RATING)
    stats1 = {metric: row1.get(metric, 0) for metric in WEIGHTS}
    stats2 = {metric: row2.get(metric, 0) for metric in WEIGHTS}
    comp1 = calculate_comp_score(stats1)
    comp2 = calculate_comp_score(stats2)
    new_r1, new_r2 = update_elo(r1, r2, comp1, comp2, outcome)
    fighter_ratings[fighter1] = new_r1
    fighter_ratings[fighter2] = new_r2
    df.at[indices[0], 'elo_after_fight'] = new_r1
    df.at[indices[1], 'elo_after_fight'] = new_r2

# ---- FINAL SAFETY NET: Fill any remaining NaNs ----
def get_current_rating(fighter):
    return fighter_ratings.get(fighter, INITIAL_RATING)

df['elo_after_fight'] = df.apply(
    lambda row: get_current_rating(row['fighter']) if pd.isna(row['elo_after_fight']) else row['elo_after_fight'],
    axis=1
)
df['elo_after_fight'] = df['elo_after_fight'].fillna(INITIAL_RATING)

# ---- CHECK ----
print("Remaining NaN values:", df['elo_after_fight'].isna().sum())
```
